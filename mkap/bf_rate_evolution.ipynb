```python
import numpy
import matplotlib.pyplot as plt
import matplotlib
import h5py
from copy import deepcopy
matplotlib.rcParams.update({'font.size': 12})
```

```python
# Loading multiple trials worth of data

interval_c7, rate_c7, se_c7 = [], [], []
interval_ar, rate_ar, se_ar = [], [], []
for i in range(1,4):
    a,b,c = numpy.load(f'{i:>02}-c7eq/bf_rate_intervals_c7eq.npy')
    interval_c7.append(a)
    rate_c7.append(b)
    se_c7.append(c)
    
    a,b,c = numpy.load(f'{i:>02}-alpha/bf_rate_intervals_c7eq.npy')
    interval_ar.append(a)
    rate_ar.append(b)
    se_ar.append(c)

interval_avg_c7, rate_avg_c7, se_avg_c7 = numpy.load('c7eq-bf_rate_intervals_c7eq.npy')
interval_avg_ar, rate_avg_ar, se_avg_ar = numpy.load('alpha-bf_rate_intervals_c7eq.npy')
```

```python
# Plotting
# Using Beta as starting state...
fig, ax = plt.subplots(ncols=1, nrows=2, figsize=(9,5), dpi=300,
                      constrained_layout=True, gridspec_kw={'height_ratios' : [20,20]}) # Use 'width_ratios' otherwise
label_list=['01','02','03','04','05']

for idx in range(0, len(interval_c7)):
    ax[0].plot(interval_c7[idx], rate_c7[idx], label=label_list[idx])
    #ax[0].fill_between(interval_c7[idx], rate_c7[idx] - se_c7[idx], y2=rate_c7[idx] + se_c7[idx], alpha=0.2)#, color='grey')
ax[0].plot(interval_avg_c7, rate_avg_c7, label='all trials')
ax[0].fill_between(interval_avg_c7, rate_avg_c7 - se_avg_c7, y2=rate_avg_c7 + se_avg_c7, alpha=0.2, color='grey')

#ax[0].set_xlabel('simulation time ($\mu s$)')
ax[0].set_ylabel('$C7_{eq}$ -> $C7_{ax}$ rate ($\mu s^{-1})$')
ax[0].legend()
ax[0].set_title('Starting from $C7_{eq}$')
#ax[0].xaxis.set_ticklabels([])

for idx in range(0, len(interval_ar)):
    ax[1].plot(interval_ar[idx], rate_ar[idx], label=label_list[idx])
    #ax[1].fill_between(interval_ar[idx], rate_ar[idx] - se_ar[idx], y2=rate_ar[idx] + se_ar[idx], alpha=0.2)#, color='grey')
ax[1].plot(interval_avg_ar, rate_avg_ar, label='all trials')
ax[1].fill_between(interval_avg_ar, rate_avg_ar - se_avg_ar, y2=rate_avg_ar + se_avg_ar, alpha=0.2, color='grey')

ax[1].set_xlabel('simulation time ($\mu s$)')
ax[1].set_ylabel('$C7_{eq}$ -> $C7_{ax}$ rate ($\mu s^{-1})$')
#ax[1].legend()
ax[1].set_title('Starting from $\\alpha_{R}$')
#fig.savefig('graphs/flux_450_3.5A6.5.pdf',dpi=300)

fig.savefig('bf_rates_c7eq.pdf',dpi=300)
```
